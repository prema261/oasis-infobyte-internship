```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

file_path = r"C:\Users\Prema\Downloads\archive (1)\Unemployment in India.csv"

# Load data
df = pd.read_csv(file_path)
df.columns = df.columns.str.strip() 

# 1. Visualization: Bar Plot (Average Unemployment by Region)
state_unemployment = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x="Estimated Unemployment Rate (%)", y="Region", data=state_unemployment, hue="Region", palette="plasma", legend=False)
plt.title("Average Unemployment Rate by Region", fontsize=14, fontweight="bold")
plt.show()

# 2. Visualization: Box Plot (Urban vs Rural)
plt.figure(figsize=(8, 5))
sns.boxplot(x="Area", y="Estimated Unemployment Rate (%)", data=df, hue="Area", palette="Set2", legend=False)
plt.title("Unemployment Distribution: Urban vs Rural", fontsize=12, fontweight="bold")
plt.show()
```

```python
# Cell 2: 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path_2 = r"C:\Users\Prema\Downloads\archive (1)\Unemployment_Rate_upto_11_2020.csv"

df2 = pd.read_csv(file_path_2)
df2.columns = df2.columns.str.strip()

# 1. Convert Date column
df2['Date'] = pd.to_datetime(df2['Date'], dayfirst=True)

# 2. Group by Date
df_trend = df2.groupby('Date')['Estimated Unemployment Rate (%)'].mean().reset_index()

# 3. Create the Line Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_trend, x="Date", y="Estimated Unemployment Rate (%)", marker='o')
plt.title("Average Unemployment Rate Over Time", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
